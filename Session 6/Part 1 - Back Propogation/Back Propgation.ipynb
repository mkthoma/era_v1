```python
import math
import matplotlib.pyplot as plt
```

```python
# Set the learning rate and the number of iterations as the epoch variable
learning_rate = 2.0
epochs = 100
```

```python
# define the targets t1 and t2
t1 = 0.5
t2= 0.5
```

```python
# define the inputs i1 and i2
i1 = 0.05
i2 = 0.1
```

```python
# define the initial set of weights and the hidden layer weights as shown in the image above
w1 = 0.15
w2 = 0.2
w3 = 0.25
w4 = 0.3
w5 = 0.4
w6 = 0.45
w7 = 0.5
w8 = 0.55
```

```python
# list of values which will be used to plot the decrease in error with each run
epoch_list = []
E_tot_list = []
```

```python
# backpropagation algorithm for a specified set of iterations
```

```python
for i in range(epochs):
    epoch_list.append(i)
    h1 = i1*w1 + i2*w2 
    a_h1 = 1/(1+math.exp(-1*h1))
    h2 = i1*w3 + i2*w4
    a_h2 = 1/(1+math.exp(-1*h2))
    o1 = w5*a_h1 + w6*a_h2
    a_o1 = 1/(1+math.exp(-1*o1))
    o2 = w7*a_h1 + w8*a_h2
    a_o2 = 1/(1+math.exp(-1*o2))

    E1 = 0.5 * (t1 - a_o1)**2
    E2 = 0.5 * (t2 - a_o2)**2
    E_tot = E1 + E2
    E_tot_list.append(E_tot)

    dw1 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w5 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w7) * a_h1 * (1 - a_h1) * i1
    dw2 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w5 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w7) * a_h1 * (1 - a_h1) * i2
    dw3 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w6 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w8) * a_h2 * (1 - a_h2) * i1
    dw4 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w6 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w8) * a_h2 * (1 - a_h2) * i2
    dw5 = (a_o1 - t1) * a_o1 * (1 - a_o1) *  a_h1
    dw6 = (a_o1 - t1) * a_o1 * (1 - a_o1) *  a_h2
    dw7 = (a_o2 - t2) * a_o2 * (1 - a_o2) *  a_h1
    dw8 = (a_o2 - t2) * a_o2 * (1 - a_o2) *  a_h2

    w1 = w1 - (learning_rate*dw1)
    w2 = w2 - (learning_rate*dw2)
    w3 = w3 - (learning_rate*dw3)
    w4 = w4 - (learning_rate*dw4)
    w5 = w5 - (learning_rate*dw5)
    w6 = w6 - (learning_rate*dw6)
    w7 = w7 - (learning_rate*dw7)
    w8 = w8 - (learning_rate*dw8)
    
```

```python
# showing error change with each iteration
plt.plot(epoch_list, E_tot_list)
plt.xlabel('Number of epochs')
plt.ylabel('Total Error')
plt.show()
```
